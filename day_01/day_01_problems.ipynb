2. How to create a series from a list, numpy array and dict?
Create a pandas series from each of the items below: a list, numpy and a dictionary


```python
import pandas as pd
import numpy as np
mylist = list('abcedfghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))

mylistSeries = pd.Series(mylist)
myarrSeries = pd.Series(myarr)
mydictSeries = pd.Series(mydict)

print(mylistSeries)
print(myarrSeries)
print(mydictSeries)
```

6. How to get the items of series A not present in series B?
Difficulty Level: L2

From ser1 remove items present in ser2.

```python
import numpy as np
import pandas as pd
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])

res = ser1[~ser1.isin(ser2)]
print(res)
```

10. How to keep only top 2 most frequent values as it is and replace everything else as ‘Other’?
Difficulty Level: L2

From ser, keep the top 2 most frequent items as it is and replace everything else as ‘Other’.

```python
import numpy as np
import pandas as pd

np.random.seed(100)
ser = pd.Series(np.random.randint(1, 5, [12]))

top2 = ser.value_counts().index[:2]

res = ser.apply(lambda x: x if x in top2 else "other")
print(res)
```

22. How to get the day of month, week number, day of year and day of week from a series of date strings?
Difficiulty Level: L2 Get the day of month, week number, day of year and day of week from

```python
import numpy as np
import pandas as pd

ser = pd.Series(['01 Jan 2010', '02-02-2011', '20120303', '2013/04/04', '2014-05-05', '2015-06-06T12:20'])

dates = pd.to_datetime(ser,format='mixed')
print(dates.dt.year)
print(dates.dt.month)
print(dates.dt.isocalendar().week)
print(dates.dt.day_name())
```

25. How to filter valid emails from a series?
Difficiulty Level: L3 Extract the valid emails from the series emails. The regex pattern for valid emails is provided as reference.

```python
import numpy as np
import pandas as pd
emails = pd.Series(['buying books at amazom.com', 'rameses@egypt.com', 'matt@t.co', 'narendra@modi.com'])
pattern ='[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\\.[A-Za-z]{2,4}'

valid_mask = emails.str.match(pattern)
valid_emails = emails[valid_mask]
print(valid_emails)
```

51. How to get the row number of the nth largest value in a column?
Difficulty Level: L2

Find the row position of the 5th largest value of column ‘a’ in df.

```python
import numpy as np
import pandas as pd
df = pd.DataFrame(np.random.randint(1, 30, 30).reshape(10,-1), columns=list('abc'))

n = 5
res = df['a'].argsort()[-n:][::-1]
print(res)
```

75. How to split a text column into two separate columns?
Difficulty Level: L2

Split the string column in df to form a dataframe with 3 columns as shown.

```python
import numpy as np
import pandas as pd
df = pd.DataFrame(["STD, City State",
"33, Kolkata West Bengal",
"44, Chennai Tamil Nadu",
"40, Hyderabad Telengana",
"80, Bangalore Karnataka"], columns=['row'])

df_out = df['row'].str.split(',', expand=True)
print(df_out)
```
